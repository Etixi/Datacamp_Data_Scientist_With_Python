####**4. Anatomie d'une docstring**

+ ***Une docstring est une chaîne écrite comme première ligne d'une fonction. Étant donné que les docstrings s'étendent généralement sur plusieurs lignes, elles sont entourées de guillemets triples, la manière de Python d'écrire des chaînes multilignes. Chaque docstring contient certaines (mais généralement pas toutes) de ces cinq informations clés : ce que fait la fonction, quels sont les arguments, quelle doit être la ou les valeurs de retour, des informations sur les erreurs générées et tout ce que vous souhaitez. à dire sur la fonction.***

####**5. Formats de chaîne de documentation**

+ ***Un style cohérent rend un projet plus facile à lire, et la communauté Python a développé plusieurs normes sur la façon de formater vos docstrings. Le style Google et Numpydoc sont les formats les plus populaires, nous allons donc nous concentrer sur ceux-ci.***

####**6. Style Google - description**

+ ***Dans le style de Google, la docstring commence par une description concise de ce que fait la fonction. Cela devrait être dans un langage impératif. Par exemple : "Split le bloc de données et empile les colonnes" au lieu de "Cette fonction divise le bloc de données et empile les colonnes".***

####**7. Style Google - arguments**

+ ***Vient ensuite la section "Args" où vous listez chaque nom d'argument, suivi de son type attendu entre parenthèses, puis de son rôle dans la fonction. Si vous avez besoin d'espace supplémentaire, vous pouvez passer à la ligne suivante et mettre en retrait comme je l'ai fait ici. Si un argument a une valeur par défaut, marquez-le comme "optionnel" lors de la description du type. Si la fonction ne prend aucun paramètre, n'hésitez pas à laisser cette section de côté.***

####**8. Style Google - valeur(s) de retour**

+ ***La section suivante est la section "Retours", où vous répertoriez le ou les types attendus de ce qui est renvoyé. Vous pouvez également fournir des commentaires sur ce qui est renvoyé, mais souvent le nom de la fonction et la description le rendront clair. Les lignes supplémentaires ne doivent pas être en retrait.***

####**9. Google-style - erreurs soulevées et notes supplémentaires**

+ ***Enfin, si votre fonction génère intentionnellement des erreurs, vous devez ajouter une section "Raises". Vous pouvez également inclure des notes supplémentaires ou des exemples d'utilisation sous forme de texte libre à la fin.***

```python
import inspect

def build_tooltip(function):
  """Create a tooltip for any function that shows the
  function's docstring.

  Args:
    function (callable): The function we want a tooltip for.

  Returns:
    str
  """
  # Get the docstring for the "function" argument by using inspect
  docstring = inspect.getdoc('__init__')
  border = '#' * 28
  return '{}\n{}\n{}'.format(border, docstring, border)
```

```python
def mean(values):
  """Get the mean of a sorted list of values

  Args:
    values (iterable of float): A list of numbers

  Returns:
    float
  """
  # Write the mean() function
  mean = sum(values) / len(values)
  return mean
```

```python
def median(values):
  """Get the median of a sorted list of values

  Args:
    values (iterable of float): A list of numbers

  Returns:
    float
  """
  # Write the median() function
  midpoint  = int(len(values) / 2)
  if len(values) % 2 == 0:
     median = (values[midpoint - 1] + values[midpoint]) / 2
  else:
   median = values[midpoint]
  return median
```

+ Ajoutez un décorateur du contextlibmodule à `la fonction timer()` qui le fera agir comme un gestionnaire de contexte.
+ Envoyez le contrôle de `la fonction timer()` au bloc de contexte.

```python
import contextlib
import time
```

```python
# Add a decorator that will make timer() a context manager
@contextlib.contextmanager
def timer():
  """Time the execution of a context block.

  Yields:
    None
  """
  start = time.time()
  # Send control back to the context block
  yield
  end = time.time()
  print('Elapsed: {:.2f}s'.format(end - start))

with timer():
  print('This should take approximately 0.25 seconds')
  time.sleep(0.25)
```

+ ***Un gestionnaire de contexte open() en lecture seule

  + prend un nom de fichier et un mode (`'r'`pour lire, `'w'`pour écrire ou `'a'`pour ajouter)
  + ouvre le fichier pour lecture, écriture ou ajout
  + rend le contrôle du contexte, avec une référence au fichier
  + attend la fin du contexte
  + puis ferme le fichier avant de quitter

+ *Votre gestionnaire de contexte fera la même chose, sauf qu'il ne prendra que le nom du fichier comme argument et qu'il n'ouvrira le fichier qu'en lecture.*

####**Instructions**

+ Contrôlez le rendement du bloc de contexte open_read_only(), en veillant à ce que l'objet read_only_file soit affecté à my_file.
+ Utilisez la méthode read_only_file de .close() pour vous assurer que vous ne laissez pas traîner des fichiers ouverts.

```python
@contextlib.contextmanager
def open_read_only(filename):
  """Open a file in read-only mode.

  Args:
    filename (str): The location of the file to read

  Yields:
    file object
  """
  read_only_file = open(filename, mode='r')
  # Yield read_only_file so it can be assigned to my_file
  yield read_only_file
  # Close read_only_file
  read_only_file.close()

with open_read_only('my_file.txt') as my_file:
  print(my_file.read())
```

+ ***C'est un gestionnaire de contexte radical en lecture seule ! Maintenant, vous pouvez vous détendre, sachant que chaque fois que vous utilisez with open_read_only() vos fichiers, ils sont protégés contre les écrasements accidentels. ####**Imprimer le type de retour**

+ ***Vous déboguez un paquet sur lequel vous avez travaillé avec vos amis. Par exemple, si vous vous attendez à ce qu'une fonction renvoie un tableau numpy, mais qu'elle renvoie une liste, vous pouvez obtenir un comportement inattendu. Pour vous assurer que ce n'est pas ce qui cause le problème, vous décidez d'écrire un décorateur, print_return_type(), qui affichera le type de la variable renvoyée à chaque appel de toute fonction qu'il décore.***

####**Instructions**

+ Créez une fonction imbriquée, wrapper(), qui deviendra la nouvelle fonction décorée.
+ Appelez la fonction en cours de décoration.
+ Renvoie la nouvelle fonction décorée.

```python
def print_return_type(func):
  # Define wrapper(), the decorated function
  def wrapper(*args, **kwargs):
    # Call the function being decorated
    result = func(*args, **kwargs)
    print('{}() returned type {}'.format(
      func.__name__, type(result)
    ))
    return result
  # Return the decorated function
  return wrapper
  
@print_return_type
def foo(value):
  return value
  
print(foo(42))
print(foo([1, 2, 3]))
print(foo({'a': 42}))
```

```python
def counter(func):
  def wrapper(*args, **kwargs):
    wrapper.count += 1
    # Call the function being decorated and return the result
    return func(*args, **kwargs)
  wrapper.count = 0
  # Return the new decorated function
  return wrapper

# Decorate foo() with the counter() decorator
@counter
def foo():
  print('calling foo()')
  
foo()
foo()

print('foo() was called {} times.'.format(foo.count))
```

```python
def bold(func):
  @wraps(func)
  def wrapper(*args, **kwargs):
    msg = func(*args, **kwargs)
    return '<b>{}</b>'.format(msg)
  return wrapper

  
def italics(func):
  @wraps(func)
  def wrapper(*args, **kwargs):
    msg = func(*args, **kwargs)
    return '<i>{}</i>'.format(msg)
  return wrapper
```

```python
def html(open_tag, close_tag):
  def decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
      msg = func(*args, **kwargs)
      return '{}{}{}'.format(open_tag, msg, close_tag)
    # Return the decorated function
    return wrapper
  # Return the decorator
  return decorator
```
