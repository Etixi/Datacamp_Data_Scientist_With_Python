####**2. Examen des types de données**

+ ***Examinons l'attribut dtypes du DataFrame. Chaque série a un type de données, qui a été automatiquement déduit par pandas lors de sa lecture dans le fichier CSV. Comme vous pouvez le voir, les seuls types de données actuellement utilisés sont object et bool. Le type de données d'objet signifie généralement qu'une série est composée de chaînes Python, bien qu'il puisse indiquer la présence d'autres objets Python tels que des listes. Le type de données bool est l'abréviation de Boolean, ce qui signifie qu'une série est composée de valeurs True et False. pandas prend également en charge d'autres types de données, tels que int pour les entiers, float pour les valeurs à virgule flottante, datetime pour les dates et les heures et category pour les variables catégorielles.***

####**3. Pourquoi les types de données sont-ils importants ?**

+ ***Mais pourquoi le type de données d'une série pandas est-il important ? Les types de données sont importants principalement parce qu'ils affectent les opérations que vous pouvez effectuer sur une série donnée. En particulier, il est avantageux de ne pas stocker les données sous forme de chaînes lorsque cela est possible. Par exemple, des opérations mathématiques peuvent être effectuées sur des entiers et des flottants, mais ces opérations échoueront si les nombres sont stockés sous forme de chaînes. Le type datetime active un riche ensemble d'attributs et de méthodes basés sur la date qui ne sont pas possibles avec des chaînes. Le type de données de catégorie entraîne une utilisation moindre de la mémoire et un traitement plus rapide que les chaînes. Et le type de données bool permet des opérations logiques et mathématiques que nous utiliserons pendant le cours.***
