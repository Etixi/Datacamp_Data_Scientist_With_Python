+ Nous allons maintenant voir comment définir des fonctions via un exemple, une fonction qui élève au carré un nombre. Le carré du nom de la fonction sera parfait pour cela.

+ Un mot rapide sur les paramètres et les arguments : lorsque vous définissez une fonction, vous écrivez des paramètres dans l'en-tête de la fonction. Lorsque vous appelez une fonction, vous transmettez des arguments à la fonction.

####**7. Docstrings**

+ Il y a un autre aspect essentiel de l'écriture de fonctions en Python : les docstrings. Les docstrings sont utilisées pour décrire ce que fait votre fonction, comme les calculs qu'elle effectue ou ses valeurs de retour. 
+ Ces descriptions servent de documentation pour votre fonction afin que quiconque lit la docstring de votre fonction comprenne ce que fait votre fonction, sans avoir à parcourir tout le code de la définition de la fonction. 
+ Les docstrings de fonction sont placés dans la ligne immédiate après l'en-tête de fonction et sont placés entre guillemets triples. Une Docstring appropriée pour notre fonction square est 'Renvoie le carré d'une valeur'.

+ Deuxièmement, alors que les listes sont définies à l'aide de crochets, les tuples sont construits à l'aide d'un ensemble de parenthèses.

+ Que se passe-t-il si nous avons une fonction interne définie dans une autre fonction externe et que nous référençons un nom x dans la fonction interne ?

+ Si Python ne trouve pas $x$ dans la portée de la fonction englobante, il recherche alors seulement la portée globale, puis la portée intégrée. Mais whoa, attendez une seconde, pourquoi imbriquons-nous même des fonctions ?

####**3. Fonctions imbriquées (2)**

+ Il y a plusieurs bonnes raisons de le faire. Disons que nous voulons utiliser un processus plusieurs fois dans une fonction. 
+ Par exemple, nous voulons une fonction qui prend 3 nombres comme paramètres et exécute la même fonction sur chacun d'eux. Une façon serait d'écrire le calcul 3 fois

####**4. Fonctions imbriquées (3)**

+ mais cela ne s'adapte certainement pas si vous devez effectuer le calcul plusieurs fois. 
+ Ce que nous pouvons faire à la place, c'est définir une fonction interne dans notre définition de fonction, comme nous le faisons ici, et l'appeler si nécessaire. 
+ C'est ce qu'on appelle une fonction imbriquée. La syntaxe de la fonction interne est exactement la même que celle de toute autre fonction.

+ N'oubliez pas non plus que l'attribution de noms ne créera ou ne modifiera que des noms locaux, à moins qu'ils ne soient déclarés dans des instructions globales ou non locales à l'aide du mot-clé global ou du mot-clé non local, respectivement.

Rappelez-vous de la vidéo précédente qui, dans la définition de la fonction, argsest un tuple.*

```python
# Define gibberish
def gibberish(*args):
    """Concatenate strings in *args together."""

    # Initialize an empty string: hodgepodge
    hodgepodge= ""

    # Concatenate the strings in args
    for word in args:
        hodgepodge += word + ' '

    # Return hodgepodge
    return hodgepodge
```

####**Instructions**

+ Complétez l'en-tête de la fonction avec le nom de la fonction report_status. Il accepte un seul argument flexible **kwargs.

```python
# Define report_status
def report_status(**kwargs):
    """Print out the status of a movie character."""

    print("\nBEGIN: REPORT\n")

    # Iterate over the key-value pairs of kwargs
    for key, value in kwargs.items():
        # Print out the keys and values, separated by a colon ':'
        print(key + ": " + value)

    print("\nEND REPORT")
```

```python
# Define count_entries()
def count_entries(df, col_name='lang'):
    """Return a dictionary with counts of
    occurrences as value for each key."""

    # Initialize an empty dictionary: cols_count
    cols_count = {}

    # Add try block
    try:
        # Extract column from DataFrame: col
        col = df[col_name]
        
        # Iterate over the column in dataframe
        for entry in col:
    
            # If entry is in cols_count, add 1
            if entry in cols_count.keys():
                cols_count[entry] += 1
            # Else add the entry to cols_count, set the value to 1
            else:
                cols_count[entry] = 1
    
        # Return the cols_count dictionary
        return cols_count

    # Add except block
    except:
        print('The DataFrame does not have a ' + col_name + ' column.')
```

```python
# Define count_entries()
def count_entries(df, col_name='lang'):
    """Return a dictionary with counts of
    occurrences as value for each key."""
    
    # Raise a ValueError if col_name is NOT in DataFrame
    if col_name not in df.columns:
        raise ValueError('The DataFrame does not have a ' + col_name + ' column.')

    # Initialize an empty dictionary: cols_count
    cols_count = {}
    
    # Extract column from DataFrame: col
    col = df[col_name]
    
    # Iterate over the column in DataFrame
    for entry in col:

        # If entry is in cols_count, add 1
        if entry in cols_count.keys():
            cols_count[entry] += 1
            # Else add the entry to cols_count, set the value to 1
        else:
            cols_count[entry] = 1
        
        # Return the cols_count dictionary
    return cols_count
```
