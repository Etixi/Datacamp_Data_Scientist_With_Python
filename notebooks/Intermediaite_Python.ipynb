####**3. et**

+ L'opérateur et fonctionne comme prévu.

+ L'opérateur ou fonctionne de la même manière, mais la différence est que seul au moins un des booléens qu'il utilise doit être vrai.

+ 5 inférieur à 7 est Vrai, 5 supérieur à 13 est Faux. L'opération or renvoie donc True.

####**5. pas**

+ Enfin, il y a l'opérateur not. Il annule simplement la valeur booléenne sur laquelle vous l'utilisez. 
+ pas vrai est faux, pas faux est vrai.

####**1. boucle while**

+ Dans le chapitre précédent, vous avez découvert la construction `if-elif-else`, un moyen de modifier le flux de vos scripts. Dès que Python rencontre une instruction if, il vérifie la condition.

####**2. si-elif-else**

+ Lorsque cette condition est vraie, le code correspondant est exécuté. Si les conditions sont évaluées à False et s'il existe des instructions elif et else, Python continue la recherche. Quoi qu'il en soit, Python ne parcourt ce morceau de code qu'une seule fois. Après cela, Python passe à la commande suivante du script. La boucle while est quelque peu similaire à une instruction if : elle exécute le code à l'intérieur si la condition est vraie. Cependant, contrairement à l'instruction if, la boucle while continuera à exécuter ce code encore et encore tant que la condition est vraie.

####**3. Pendant que**

+ La syntaxe d'une boucle while est très similaire à l'instruction if, comme vous pouvez le voir ici. La boucle while n'est pas si courante, mais dans certains cas, elle peut être très utile. Par exemple, supposons que vous calculiez numériquement un modèle basé sur vos données. Cela implique généralement de répéter les mêmes étapes encore et encore, jusqu'à ce que l'erreur entre votre modèle et vos données soit inférieure à une certaine limite. Lorsque vous pouvez reformuler le problème en 'répéter une action jusqu'à ce qu'une condition particulière soit remplie', une boucle while est souvent la solution. Supposons que nous commencions avec une erreur de cinquante et que notre algorithme fantaisiste divise l'erreur par quatre à chaque exécution. Nous voulons continuer jusqu'à ce que l'erreur ne soit plus supérieure à 1.

####**4. Pendant que**

+ Passons maintenant au script, nous commençons par ajouter l'erreur avec laquelle nous commençons, 50. Ensuite, nous écrivons une boucle while. Dans la partie condition, nous écrivons erreur > 1, de sorte que la boucle while s'exécute à nouveau tant que l'erreur est supérieure à 1. Dans le code, nous divisons l'erreur par quatre et mettons à jour la variable d'erreur. Ce type de simule notre algorithme sophistiqué qui divise l'erreur par quatre à chaque exécution. Ensuite, nous imprimons également cette erreur. Passons en revue ce qui se passe si vous exécutez réellement ce script, étape par étape.

####**6. Pendant que**

+ Maintenant, Python revient à la condition de la boucle while, avec une erreur égale à 12,5. Encore une fois, la condition est vraie et le code est exécuté. L'erreur est divisée par 4 et imprimée. Maintenant, l'erreur est 3.125 et Python revient à la condition while.
