+ Calculez le poids moyen des bébés nés à terme.

```python
def resample_rows_weighted(df, column='wgt2013_2015'):
    """Resamples a DataFrame using probabilities proportional to given column.
    Args:
        df: DataFrame
        column: string column name to use as weights
    returns: 
        DataFrame
    """
    weights = df[column].copy()
    weights /= sum(weights)
    indices = np.random.choice(df.index, len(df), replace=True, p=weights)
    sample = df.loc[indices]
    return sample
```

```python
# Resample the data
nsfg = resample_rows_weighted(nsfg, 'wgt2013_2015')

# Clean the weight variables
pounds = nsfg['birthwgt_lb1'].replace([98, 99], np.nan)
ounces = nsfg['birthwgt_oz1'].replace([98, 99], np.nan)

# Compute total birth weight
birth_weight = pounds + ounces/16
```

####**1. Fonctions de distribution cumulées**

+ ***Dans la leçon précédente, nous avons vu la fonction de masse de probabilité, ou PMF, qui représente les valeurs possibles dans une distribution et leurs probabilités. Dans cette leçon, nous verrons une autre façon de représenter une distribution, `la fonction de distribution cumulative ou CDF`. Les CDF sont utiles pour certains calculs ; ils sont également un excellent moyen de visualiser et de comparer les distributions.***

####**2. Du PMF au CDF**

+ ***Vous vous souvenez peut-être qu'un PMF vous indique - si vous tirez une valeur aléatoire d'une distribution - quelle est la chance d'obtenir $x$, pour toute valeur donnée de $x$. Le CDF est similaire; si vous tirez une valeur aléatoire d'une distribution, cela vous indique la chance d'obtenir une valeur inférieure ou égale à $x$.***

####**3. Exemple**

+ ***Comme petit exemple, supposons que la distribution ne comporte que 5 éléments, 1, 2, 2, 3 et 5. Le PMF indique que la probabilité de la valeur 1 est de 1/5 ; la probabilité de la valeur 2 est de 2/5 et les probabilités de 3 et 5 sont de 1/5. Le CDF est la somme cumulée des probabilités du PMF. Par exemple, la CDF de 2 est de trois cinquièmes, car trois valeurs sur 5 dans la distribution sont inférieures ou égales à 2. La CDF de 5 est 1, ou 100 %, car toutes les valeurs de la distribution sont inférieures à ou égal à 5.***

####**4. Créer et tracer un CDF**

+ ***Dans le code de ce cours, je fournis une classe Cdf qui est similaire à la classe Pmf que nous avons vue. En tant que fonction, Cdf prend n'importe quel type de séquence et renvoie un nouvel objet Cdf. Dans cet exemple, la séquence correspond aux âges des répondants de l'Enquête sociale générale. Le Cdf fournit plot, qui trace le CDF sous forme de ligne. Voici à quoi ça ressemble. L'axe des x représente les âges, de 18 à 89 ans. L'axe des $y$ représente les probabilités cumulées, de 0 à 1.***

####**5. Évaluer le CDF**

+ ***L'objet Cdf peut être utilisé comme une fonction, donc si vous lui donnez un âge, il renvoie la probabilité correspondante. Dans cet exemple, l'âge est la quantité, q, qui est 51. La probabilité correspondante est p, qui est 0,66. Cela signifie qu'environ 66% des répondants ont 51 ans ou moins. La flèche sur la figure montre comment vous pourriez lire cette valeur à partir du CDF, au moins approximativement.***

####**6. Évaluer le CDF inverse**

+ ***La CDF est une fonction inversible, ce qui signifie que si vous avez une probabilité, p, vous pouvez rechercher la quantité correspondante, q. Dans cet exemple, je recherche la probabilité 0,25, qui renvoie 30. Cela signifie que 25 % des répondants ont 30 ans ou moins. Une autre façon de dire la même chose est "l'âge de 30 ans est le 25e centile de cette distribution". Je recherche également la probabilité 0,75, qui renvoie 57, donc 75 % des répondants ont 57 ans ou moins. Encore une fois, les flèches de la figure montrent comment vous pouvez lire ces valeurs à partir du CDF. Soit dit en passant, la distance entre le 25e et le 75e centile s'appelle l'intervalle interquartile, ou IQR. Il mesure la propagation de la distribution, il est donc similaire à l'écart type ou à la variance. Parce qu'il est basé sur des centiles, il n'est pas ébranlé par des valeurs extrêmes ou des valeurs aberrantes, comme le fait la variance. Ainsi, l'IQR peut être plus "robuste" que la variance, ce qui signifie qu'il fonctionne bien même s'il y a des erreurs dans les données ou des valeurs extrêmes.***

+ ***Quand les gens disent "corrélation" avec désinvolture, ils peuvent signifier n'importe quelle relation entre deux variables. En statistique, cela signifie généralement le coefficient de corrélation de Pearson, qui est un nombre compris entre -1 et 1 qui quantifie la force d'une relation linéaire entre les variables. Pour illustrer, je vais sélectionner trois colonnes de l'ensemble de données BRFSS, comme ceci. Le résultat est un DataFrame avec uniquement ces colonnes. ####**5. Relations non linéaires**

+ ***Mais la corrélation ne fonctionne que pour les relations linéaires. Si la relation n'est pas linéaire, la corrélation sous-estime généralement sa force. Pour démontrer, je vais générer de fausses données : xs contient des points équidistants entre -1 et 1. ys est xs au carré plus un bruit aléatoire. Voici le nuage de points de xs et ys. Mais voici la matrice de corrélation ; la corrélation calculée est proche de 0. En général, si la corrélation est élevée, c'est-à-dire proche de 1 ou -1, vous pouvez en conclure qu'il existe une forte relation linéaire. Mais si la corrélation est proche de 0, cela ne signifie pas qu'il n'y a pas de relation ; il pourrait y avoir une forte relation non linéaire.***
