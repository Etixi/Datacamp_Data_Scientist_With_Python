####**Ajustement du nombre de cases dans un histogramme**


+ ***L'histogramme que vous venez de créer comportait dix cases. C'est la valeur par défaut de matplotlib. La « règle de la racine carrée » est une règle empirique couramment utilisée pour choisir le nombre de bacs : choisissez le nombre de bacs comme étant la racine carrée du nombre d'échantillons. Tracez à nouveau l'histogramme des longueurs de pétales d'Iris versicolor , cette fois en utilisant la règle de la racine carrée pour le nombre de bacs. Vous spécifiez le nombre de bacs à l'aide de l' binsargument mot-clé of plt.hist().***

+ *Les utilitaires de traçage sont déjà importés et les valeurs par défaut de Seaborn sont déjà définies. La variable versicolor_petal_length contient un tableau de longueurs de pétales et se trouve déjà dans votre espace de noms.*

####**Instructions**

+ Importer numpyau format np. Cela donne accès à la fonction racine carrée, np.sqrt().
+ Déterminez le nombre de points de données dont vous disposez à l'aide de len().
+ Calculez le nombre de bacs en utilisant la règle de la racine carrée.
+ Convertissez le nombre de bacs en nombre entier à l'aide de la fonction intégrée int().
+ Générez l'histogramme et assurez-vous d'utiliser l'argument bins de mot-clé.
+ Cliquez sur Soumettre pour tracer la figure et voir le fruit de votre travail !

```python
# Import numpy
import numpy as np

# Compute number of data points: n_data
n_data=len(versicolor_petal_length)

# Number of bins is the square root of number of data points: n_bins
n_bins=np.sqrt(n_data)

# Convert number of bins to integer: n_bins
n_bins=int(n_bins)

# Plot the histogram
_=plt.hist(versicolor_petal_length, n_bins)

# Label axes
_ = plt.xlabel('petal length (cm)')
_ = plt.ylabel('count')

# Show histogram
plt.show()
```

####**Calcul de l'ECDF**

+ ***Dans cet exercice, vous allez écrire une fonction qui prend en entrée un tableau 1D de données, puis renvoie les valeurs xet yde l'ECDF. Vous utiliserez cette fonction encore et encore tout au long de ce cours et de sa suite. Les ECDF sont parmi les parcelles les plus importantes dans l'analyse statistique. Vous pouvez écrire votre propre fonction, `foo(x,y)` selon le squelette suivant :***

      def foo(a,b):
        """State what function does here"""
        # Computation performed here
        return x, y

+ ***La fonction `foo()`ci-dessus prend deux arguments $a$ et $b$ et renvoie deux valeurs $x$ et $y$. L'en-tête de la fonction `def foo(a,b)`:contient la signature de la fonction `foo(a,b)`, qui se compose du nom de la fonction, ainsi que de ses paramètres. Pour en savoir plus sur l'écriture de vos propres fonctions, consultez [le cours de DataCamp Python Data Science Toolbox (Part 1)]() ici !***

####**Instructions**

+ Définissez une fonction avec la signature ecdf(data). Dans la définition de la fonction,
+ Calculez le nombre de points de données, $n$, à l'aide de la len()fonction.
+ les valeurs $x$ sont les données triées. Utilisez la np.sort()fonction pour effectuer le tri.
+ les données $y$ de l'ECDF vont de 1/nà par 1incréments équidistants. Vous pouvez le construire en utilisant np.arange().

+ N'oubliez pas, cependant, que la valeur finale dans np.arange()n'est pas inclusive. Par conséquent, np.arange() il faudra passer de $1$ à $n+1$. Assurez-vous de diviser cela par $n$.
+ La fonction renvoie les valeurs $x$ et $y$.

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n= len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

####**Calcul du coefficient de corrélation de Pearson**

+ ***Comme mentionné dans la vidéo, le coefficient de corrélation de Pearson, également appelé Pearson r, est souvent plus facile à interpréter que la covariance. Il est calculé à l'aide de la np.corrcoef()fonction. Comme np.cov(), il prend deux tableaux comme arguments et renvoie un tableau 2D. Les entrées [0,0] et [1,1]sont nécessairement égales à 1(savez-vous pourquoi ?), et la valeur que nous recherchons est entrée [0,1].***

+ Définir une fonction avec signature pearson_r($x, y$).
+ Utilisez np.corrcoef() pour calculer la matrice de corrélation de $x$ et $y$(passez-les à np.corrcoef()dans cet ordre).
+ La fonction renvoie l'entrée [0,1] de la matrice de corrélation.
+ Calculez la corrélation de Pearson entre les données des tableaux versicolor_petal_length et versicolor_petal_width.
+ Affectez le résultat à r.
+ Imprimez le résultat.


```python
def pearson_r(x, y):
    """Compute Pearson correlation coefficient between two arrays."""
    # Compute correlation matrix: corr_mat
    corr_mat=np.corrcoef(x,y)

    # Return entry [0,1]
    return corr_mat[0,1]

# Compute Pearson correlation coefficient for I. versicolor: r
r=pearson_r(versicolor_petal_length,versicolor_petal_width)

# Print the result
print(r)
```

####**Instructions**

+ Définir une fonction avec signature perform_bernoulli_trials($n, p$).
+ Initialiser à zéro une variable n_successle compteur de Trues, qui sont des succès d'essai de Bernoulli.
+ Écrivez une forboucle dans laquelle vous effectuez un essai de Bernoulli à chaque itération et incrémentez le nombre de succès si le résultat est True. Effectuez ndes itérations en bouclant sur range(n).
+ Pour effectuer un essai de Bernoulli, choisissez un nombre aléatoire entre zéro et un à l'aide de rng.random().
+ Si le nombre que vous avez choisi est inférieur à p, incrémentez n_success(utilisez l' += 1opérateur pour y parvenir). Un RNG a déjà été instancié en tant que variable rnget amorcé.
+ La fonction renvoie le nombre de succès n_success.

```python
def perform_bernoulli_trials(n, p):
    """Perform n Bernoulli trials with success probability p
    and return number of successes."""
    # Initialize number of successes: n_success
    n_success = 0

    # Perform trials
    for i in range(n):
        # Choose random number between zero and one: random_number
        random_number = rng.random()

        # If less than p, it's a success so add one to n_success
        if random_number < p:
            n_success += 1

    return n_success
```

####**Le CDF normal**

+ ***Maintenant que vous avez une idée de l'apparence du PDF normal, considérons son CDF. À l'aide des exemples que vous avez générés dans le dernier exercice (dans votre espace de noms en tant que `samples_std1, samples_std3 et samples_std10`), générez et tracez les `CDF`.***

####**Instructions**

+ Utilisez votre fonction ecdf() pour générer des valeurs $x$ et $y$ pour les CDF : `x_std1, y_std1, x_std3, y_std3 et x_std10, y_std10`, respectivement.
+ Tracez les trois CDF sous forme de points (n'oubliez pas les arguments mot-clé markeret !).linestyle
+ Cliquez sur Soumettre pour créer une légende indiquant les écarts-types que vous avez utilisés et pour afficher votre graphique.
+ Il n'est pas nécessaire d'étiqueter les axes car nous n'avons pas défini ce qui est décrit par la distribution normale ; nous examinons simplement les formes des CDF.

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n + 1) / n

    return x, y
```

```python
# Generate CDFs
x_std1, y_std1=ecdf(samples_std1)
x_std3, y_std3=ecdf(samples_std3)
x_std10, y_std10= ecdf(samples_std10)

# Plot CDFs
_ = plt.plot(x_std1, y_std1, marker='.', linestyle='none')
_ = plt.plot(x_std3, y_std3, marker='.', linestyle='none')
_ = plt.plot(x_std10, y_std10, marker='.', linestyle='none')


# Make a legend and show the plot
_ = plt.legend(('std = 1', 'std = 3', 'std = 10'), loc='lower right')
plt.show()
```

***Bon travail! Les CDF passent tous par la moyenne au 50e centile ; la moyenne et la médiane d'une distribution normale sont égales. La largeur de la CDF varie avec l'écart type.***

+ Dessinez les temps d'attente (`size`: nombre d'échantillons) pour atteindre le cycle à partir d'une distribution exponentielle paramétrée par tau2 et attribuez-la à $t2$.
+ La fonction renvoie la somme des temps d'attente pour les deux événements.

```python
def successive_poisson(tau1, tau2, size=1):
    """Compute time for arrival of 2 successive Poisson processes."""
    # Draw samples out of first exponential distribution: t1
    t1 = rng.exponential(tau1, size=size)

    # Draw samples out of second exponential distribution: t2
    t2 = rng.exponential(tau2, size=size)

    return t1 + t2
```
