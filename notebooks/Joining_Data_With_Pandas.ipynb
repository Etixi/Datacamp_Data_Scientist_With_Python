####**1. Relations un à plusieurs**

+ Content de te revoir! Dans la dernière leçon, nous avons appris à fusionner deux DataFrames avec la méthode de fusion. Dans cette leçon, nous aborderons différents types de relations entre les tables. En particulier, nous discuterons de la relation un-à-plusieurs. Mais d'abord, considérons rapidement ce qu'est une relation un à un.

####**2. Individuel**

+ Dans une relation un-à-un, chaque ligne du tableau de gauche est liée à une et une seule ligne du tableau de droite.

####**3. Exemple individuel**

+ Nous avons examiné une relation un à un plus tôt. 

+ Rappelez-vous la relation entre le tableau des quartiers et le tableau du recensement. 

+ Chaque ligne du tableau des quartiers est liée à une seule ligne du tableau de recensement, il n'y a donc qu'une seule ligne pour le quartier 3 dans chaque tableau. En pratique, il est logique qu'il y ait une ligne d'informations sur la population pour chaque quartier. 
+ Cela n'aurait aucun sens si le tableau de recensement contenait plusieurs valeurs de population en 2000 pour le troisième quartier.

####**. Un à plusieurs**

+ Alors, qu'est-ce qu'une relation un-à-plusieurs ? Eh bien, dans une relation un-à-plusieurs, chaque ligne de la table de gauche est liée à une ou plusieurs lignes de la table de droite.

####**5. Exemple un-à-plusieurs**

+ Pour donner un exemple d'une relation un-à-plusieurs, revenons à notre table de quartiers. Dans chaque quartier, il y a de nombreuses entreprises. Nous fusionnerons le tableau des quartiers avec un tableau des entreprises agréées dans chaque quartier.

####**8. Exemple un-à-plusieurs**

+ orsque nous fusionnons les deux tables avec la méthode de fusion, en définissant l'attribut "on" sur la colonne ward, la table résultante contient à la fois des données de ward locales et des données de licence commerciale. 
+ Notez que le quartier 1 et son échevin Joe sont répétés dans le tableau résultant car le tableau des licences contient de nombreuses entreprises dans le 1er quartier.
