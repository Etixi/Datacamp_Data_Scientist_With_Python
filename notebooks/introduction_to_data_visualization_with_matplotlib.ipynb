####**Création de petits multiples avec plt.subplots**

+ De petits multiples sont utilisés pour tracer plusieurs ensembles de données côte à côte. Dans Matplotlib, de petits multiples peuvent être créés à l'aide de la fonction plt.subplots(). 
+ Le premier argument est le nombre de lignes dans le tableau d'objets Axes générés et le second argument est le nombre de colonnes. Dans cet exercice, vous utiliserez les données d'Austin et de Seattle pour vous entraîner à créer et remplir un tableau de sous-parcelles.

####**Instructions**

####**Définition d'une fonction qui trace des données de séries chronologiques**

+ Une fois que vous réalisez qu'une section particulière de code que vous avez écrite est utile, il est judicieux de définir une fonction qui enregistre cette section de code pour vous, plutôt que de la copier dans d'autres parties de votre programme où vous souhaitez utiliser ce code.

+ Ici, nous allons définir une fonction qui prend des entrées telles qu'une variable de temps et une autre variable et les trace sous forme d'entrées $x$ et $y$. Ensuite, il définit les étiquettes sur les axes $x$ et $y$ et définit les couleurs de l'étiquette de l'axe $y$, les graduations de l'axe y et les étiquettes de graduation.

####**Instructions**

+ Définissez une fonction appelée plot_time series qui prend en entrée un objet Axes ( axes), des données ($x$, $y$), une chaîne avec le nom d'une couleur et des chaînes pour les étiquettes des axes $x$ et $y$.
+ Tracez $y$ en fonction de dans la couleur fournie en entrée color.
+ Définissez les étiquettes des axes x et y à l'aide de l'entrée fournie xlabel et ylabel, définissez la couleur de l'étiquette de l'axe $y$ à l'aide de color.
+ Définissez les paramètres de graduation de l'axe $y$ à l'aide de la méthode tick_params de l'objet Axes, en définissant le mot-clé colors sur color.

```python
# Define a function called plot_timeseries
def plot_timeseries(axes, x, y, color, xlabel, ylabel):

  # Plot the inputs x,y in the provided color
  axes.plot(x, y, color=color)

  # Set the x-axis label
  axes.set_xlabel(xlabel)

  # Set the y-axis label
  axes.set_ylabel(ylabel, color=color)

  # Set the colors tick params for y-axis
  axes.tick_params('y', colors=color)
```

+ L'une des principales forces de Matplotlib est qu'il peut être automatisé pour s'adapter aux données qu'il reçoit en entrée. Par exemple, si vous recevez des données contenant un nombre inconnu de catégories, vous pouvez toujours créer un diagramme à barres comportant des barres pour chaque catégorie.

+ C'est ce que vous ferez dans cet exercice. Vous visualiserez à nouveau des données sur les médaillés des Jeux olympiques d'été de 2016, mais cette fois, vous disposerez d'un ensemble de données contenant un nombre inconnu de branches sportives. Celui-ci sera chargé en mémoire sous la forme d'un DataFrameobjet pandas appelé summer_2016_medals, qui a une colonne appelée "Sport"qui vous indique à quelle branche du sport correspond chaque ligne. Il y a aussi une "Weight"colonne qui vous indique le poids de chaque athlète.

####**Instructions**

+ Itérez sur les valeurs de sportsparamètre sporten tant que variable de boucle.
+ À chaque itération, extrayez les lignes où la "Sport"colonne est égale à sport.
+ Ajoutez une barre à l'objet ax fourni, étiquetée avec le nom du sport, avec la moyenne de la "Weight"colonne comme hauteur et l'écart type comme barre d'erreur sur l'axe y.
+ Enregistrez la figure dans le fichier "sports_weights.png".

```python
fig, ax = plt.subplots()

# Loop over the different sports branches
for sport in sports:
  # Extract the rows only for this sport
  sport_df = summer_2016_medals[summer_2016_medals["Sport"]==sport]
  # Add a bar for the "Weight" mean with std y error bar
  ax.bar(sport, sport_df["Weight"].mean(), yerr=sport_df["Weight"].std())

ax.set_ylabel("Weight")
ax.set_xticklabels(sports, rotation=90)

# Save the figure to file
fig.savefig("sports_weights.png")
```
