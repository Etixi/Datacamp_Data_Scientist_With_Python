+ ***Nous avons créé trois listes de données à partir de cet ensemble de données pour vous aider à démarrer. gdpest une liste qui contient la valeur du PIB par pays, exprimée en dollars par personne. phonesest une liste du nombre de téléphones portables pour 1 000 habitants dans ce pays. Enfin, percent_literate est une liste qui contient le pourcentage de la population de chaque pays qui sait lire et écrire.***

####**Consignes**

+ Importez Matplotlib et Seaborn en utilisant la convention de dénomination standard.
+ Créez un nuage de points du PIB (gdp) par rapport au nombre de téléphones pour 1 000 personnes (phones).

```python
# Import Matplotlib, Pandas and Seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (10, 5)
```

```python
df = pd.read_csv('/content/countries-of-the-world.csv')
df.head()
```

```python
gdp = df['GDP ($ per capita)'].tolist()
phones = df['Phones (per 1000)'].tolist()
percent_literate = df['Literacy (%)'].tolist()
region = df['Region'].tolist()
```

```python
df.head()
```

```python
# Create scatter plot with GDP on the x-axis and number of phones on the y-axis
sns.scatterplot(x=gdp, y=phones)

# Show plot
plt.show()
```

```python
# Change this scatter plot to have percent literate on the y-axis
sns.scatterplot(x=gdp, y=percent_literate)

# Show plot
plt.show()
```

***Bien! Bien que ce graphique ne montre pas de relation linéaire entre le PIB et le pourcentage d'alphabétisés, les pays ayant un PIB inférieur semblent plus susceptibles d'avoir un pourcentage inférieur de la population sachant lire et écrire.***
