####**2. Format ISO 8601**

+ ***Par exemple, créons une date et voyons comment Python l'imprime par défaut. Comme précédemment, nous importons date depuis datetime et créons à nouveau un objet pour le 5 novembre 2017. Lorsque nous demandons à Python d'imprimer la date, il imprime l'année, le jour puis le mois, séparés par des tirets, et toujours avec deux chiffres pour le jour et le mois. Dans le commentaire, vous pouvez voir que j'ai noté cela comme AAAA-MM-JJ ; année à quatre chiffres, mois à deux chiffres et jour du mois à deux chiffres. Ce format par défaut est également appelé format ISO, ou format ISO 8601, d'après la norme internationale ISO 8601 sur laquelle il est basé. Les chaînes ISO 8601 ont toujours la même longueur puisque le mois et le jour sont écrits avec des 0 lorsqu'ils sont inférieurs à 10. Nous parlerons d'un autre avantage de l'ISO 8601 dans un instant. Si nous voulons la représentation ISO d'une date sous forme de chaîne, disons de l'écrire dans un fichier CSV au lieu de simplement l'imprimer, vous pouvez appeler la méthode `isoformat()`. Dans cet exemple, nous l'avons placé dans une liste afin que vous puissiez voir qu'il crée une chaîne.***

####**3. Format ISO 8601**

+ ***Le format ISO 8601 présente un autre avantage intéressant. Pour illustrer, nous avons créé une variable appelée some_dates et représenté ici deux dates sous forme de chaînes : 1er janvier 2000 et 31 décembre 1999. Les dates au format ISO 8601 sont triées correctement. Lorsque nous imprimons la version triée de cette liste, le premier jour est le premier et le plus tardif est le deuxième. Par exemple, si nous utilisons des dates ISO 8601 dans les noms de fichiers, elles peuvent être correctement triées du plus ancien au plus récent. Si nous avions le mois ou le jour en premier, les chaînes ne seraient pas triées par ordre chronologique.***

####**4. Tous les autres formats**

+ ***Si vous ne souhaitez pas mettre les dates au format ISO 8601, Python dispose d'un ensemble flexible d'options pour représenter les dates d'autres manières, en utilisant `la méthode strftime()`.***

####**5. Tous les autres formats : strftime**

+ ***`strftime()` fonctionne en vous permettant de passer une "chaîne de format" que Python utilise pour formater votre date. Voyons un exemple. Nous créons à nouveau un exemple de date du 5 janvier 2017. Nous appelons ensuite `strftime()` sur d, avec la chaîne de format de % capital Y. Strftime lit le % capital Y et remplit l'année dans cette chaîne pour nous. `strftime()` est cependant très flexible : nous pouvons lui donner des chaînes arbitraires avec % majuscule Y pour la chaîne de format, et il collera l'année. Par exemple, nous pouvons utiliser la chaîne de format "L'année est %Y".***

####**6. Tous les autres formats : strftime**

+ ***`Strftime` a d'autres espaces réservés en plus de %Y : % minuscule m donne le mois, et % minuscule d donne le jour du mois. En les utilisant, nous pouvons représenter des dates dans des formats arbitraires pour tous nos besoins.***

####**Nettoyage des données d'heure d'été avec pli***


+ ***Comme nous venons de le découvrir, il y a un trajet dans notre ensemble de données qui est gâché par un passage à l'heure d'été. Nettoyons l'ensemble de données afin d'avoir une longueur de trajet minimale correcte. Nous pouvons utiliser le fait que nous savons que la fin du trajet s'est produite après le début pour corriger la durée gâchée par le passage à l'heure d'été.***

+ ***Étant donné que Python ne gère pas l'arithmétique tz.enfold(), nous devons placer nos objets datetime dans UTC, où les ambiguïtés ont été résolues.***

+ *onebike_datetimes est déjà chargé et dans le bon fuseau horaire. tzet timezone ont été importés. À utiliser tz.UTC pour le fuseau horaire.*

####**Instructions**

+ Complétez l'énoncé if pour qu'il soit vrai uniquement lorsqu'un trajet start vient après son end.
+ Quand start est après end, appelez tz.enfold() le end afin que vous sachiez qu'il fait référence à celui qui suit le changement d'heure d'été.
+ Après l'instruction if, convertissez le début et la fin en UTC afin de pouvoir effectuer une comparaison appropriée.



```python
trip_durations = []
for trip in onebike_datetimes:
  # When the start is later than the end, set the fold to be 1
  if trip['start'] > trip['end']:
    trip['end'] = tz.enfold(trip['end'])
  # Convert to UTC
  start = trip['start'].astimezone(tz.UTC)
  end = trip['end'].astimezone(tz.UTC)

  # Subtract the difference
  trip_length_seconds = (end-start).total_seconds()
  trip_durations.append(trip_length_seconds)

# Take the shortest trip duration
print("Shortest trip: " + str(min(trip_durations)))
```

Alors que .resample()regroupe les lignes selon des informations d'heure ou de date, .groupby()regroupe les lignes en fonction des valeurs d'une ou de plusieurs colonnes. + *.resample() peut être appelé après .groupby(). Par exemple, combien de temps a duré le trajet médian par mois et par type d'abonnement ?*

```python
# Group rides by member type, and resample to the month
grouped = rides.groupby('Member type')\
  .resample('M', on='Start date')

# Print the median duration for each group
print(grouped['Duration'].median())
```

####**Combien de temps entre les trajets ?**

+ ***Pour votre dernier exercice, profitons de l'indexation Pandas pour faire quelque chose d'intéressant. Combien de temps s'est-il écoulé entre les trajets ?***

####**Instructions**

+ Calculez la différence entre la Start datede la ligne actuelle et la End datede la ligne précédente et affectez-la à rides['Time since'].

+ Convertir rides['Time since'] en secondes pour faciliter le travail.
+ Rééchantillonner ridespour être dans des compartiments mensuels selon le Start date.
+ Divisez la moyenne par (60*60) pour obtenir le nombre d'heures en moyenne pendant lesquelles W20529 a attendu sur le quai avant d'être repris.



```python
rides['Start date'] = rides['Start date'].dt.tz_convert('America/New_York')
rides['End date'] = rides['End date'].dt.tz_localize('America/New_York', ambiguous='NaT')
```

```python
# Shift the index of the end date up one; now subract it from the start date
rides['Time since'] = rides['Start date'] - (rides['End date'].shift(1))

# Move from a timedelta to a number of seconds, which is easier to work with
rides['Time since'] = rides['Time since'].dt.total_seconds()

# Resample to the month
monthly = rides.resample('M', on='Start date')

# Print the average hours between rides each month
print(monthly['Time since'].mean()/(60*60))
```
